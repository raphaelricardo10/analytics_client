```python
load_dotenv("./env")

bq_client = BigQueryClient(
    project_id=os.getenv("PROJECT_ID"), dataset=os.getenv("BQ_DATASET")
)
```

```python
df = bq_client.select(
    "multisensor_data",
    to_df=True,
    limit=None,
).drop_duplicates()

df.head()
df = df.set_index("ts")
```

```python
df.index = pd.to_datetime(df.index,unit='s')
```

```python
subset = df[df.device == '1c:bf:ce:15:ec:4d'].drop(['device'], axis=1).head(50)
corr = pd.DataFrame(1 - squareform(pdist(subset, lambda u,v: (u != v).mean())))
corr.describe()
```
